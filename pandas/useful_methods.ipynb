```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
df['tip'] = df['tip'].apply(lambda t: t +1)
```

```python
df.head()
```

```python
df['CC Number'] = df['CC Number'].apply(lambda cc: str(cc)[-4:])
```

```python
df.head()
```

```python
df.info()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >=10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```
