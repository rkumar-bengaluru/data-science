```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
sns.countplot(x='work experience',data=df)
```

```python
sns.countplot(x='division',data=df)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df,hue='training level')
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.barplot(x='level of education',y='sales', data=df,estimator=np.mean)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.barplot(x='level of education',y='salary', data=df,estimator=np.mean)
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.barplot(x='level of education',y='salary', data=df,estimator=np.mean,hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
text = ['Advanced antivirus software defend cyberthreats ',
        'people living in groups make decisions',
        'extraordinary year australian films']
coun_vect = CountVectorizer()
count_matrix = coun_vect.fit_transform(text)
count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = coun_vect.get_feature_names(),index=['software_doc','people_doc','film_doc'])
df.head()
```

```python
columns = ['a','b','c','d']
y = [[1,1,1,'software_doc'],
     [1,1,1,'people_doc'],
     [1,1,1,'film_doc'],
    [1,1,1,'software_doc'],
    [1,1,1,'people_doc'],
    [1,1,1,'people_doc'],
    [1,1,1,'film_doc']]
df = pd.DataFrame(data=y,columns=columns)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=df,x='d')
plt.xlabel('topics')
plt.ylabel('frequency')
```
