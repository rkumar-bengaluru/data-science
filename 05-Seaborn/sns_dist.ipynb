```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# displot
# rugplot
# kdeplot
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
#plt.figure(figsize=(8,5),dpi=200)
sns.rugplot(x='salary',height=.5,data=df)
```

```python
#sns.set(style='darkgrid')
sns.displot(x='salary',data=df,bins=20)
```

```python
sns.histplot(x='salary',data=df)
```

```python
sns.displot(x='salary',data=df,bins=20,kde=True,rug=True)
```

```python
sns.kdeplot(x='salary',data=df)
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
sample_ages.head()
```

```python
sns.rugplot(x='age',data=sample_ages)
```

```python
sns.displot(x='age', data=sample_ages)
```

```python
sns.displot(x='age',data=sample_ages,kde=True,rug=True)
```

```python
sns.kdeplot(x='age',data=sample_ages,clip=[0,100])
```

```python
sns.kdeplot(x='age',data=sample_ages,clip=[0,100],bw_adjust=0.03)
```

```python
sns.kdeplot(x='age',data=sample_ages,clip=[0,100],bw_adjust=0.03,shade=True)
```
