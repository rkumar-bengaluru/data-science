```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

<h2>Gradient Descent in 1 D</h2>

```python
#y = 6*x - 3
#0 = 6*x - 3
#6x = 3
#x= 3/6 = 1/2 = 0.5
```

```python
# function (fx) = 3x^2 - 3x + 4
def fx(x):
    return 3*x**2 - 3*x + 4
```

```python
# derivative function f^(x) = 6x - 3
def derivative_fx(x):
    return 6*x - 3
```

```python
# plot the function and its derivatives
x = np.linspace(-2,2,2001)

# plotting
plt.plot(x,fx(x),x,derivative_fx(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x,1)
print(localmin)
# learning params
learning_rate = .01
training_epochs = 100
# run through training
for i in range(training_epochs):
    grad = derivative_fx(localmin)
    localmin = localmin - learning_rate * grad
    
localmin
```

```python
# plot the results
plt.plot(x,fx(x),x,derivative_fx(x))
plt.plot(localmin,derivative_fx(localmin),'ro')
plt.plot(localmin,fx(localmin),'ro')
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum : %s'%localmin[0])
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x,1)
print(localmin)
# learning params
learning_rate = .01
training_epochs = 100
# run through training and store all results
modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
    grad = derivative_fx(localmin)
    localmin = localmin - learning_rate * grad
    modelparams[i,:] = localmin,grad
    
print(localmin)
#print(modelparams)
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    ax[i].plot(modelparams[:,i],'o-')
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'Final Estimated minimum : {localmin[0]:.5f}')

ax[0].set_ylabel('Local Minimum')
ax[1].set_ylabel('Derivative')
plt.show()
```

```python
import math
# function (fx) = cos(2pix) + x^2 , x in (-2,2)
def fx(x):
    a = np.cos(2*math.pi*x)
    b = x**2
    return (a +b)

# f^(x) = -sin(2*pi*x)*2pi + 2x
def der_fx(x):
    a = -(np.sin(2*math.pi*x))*2*math.pi
    b = 2*x
    return (a +b)
```

```python
# plot the function and its derivatives
x = np.linspace(-2,2,2001)

# plotting
plt.plot(x,fx(x),x,der_fx(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```
