```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='Linear')
ax.plot(x,x**2,label='Exponential')
ax.legend(loc='upper left')
```

```python
x = np.linspace(0,10,11)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='red',label='x v/s x',lw=2,ls='--')
line = ax.plot(x,x*2,color='yellow',label='x v/s x^2',lw=3,ls='-.',marker='o')
line[0].set_dashes([1,1,1,2])
ax.legend()
```
