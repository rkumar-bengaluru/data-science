```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import textwrap

from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
nltk.download('stopwords')
```

```python
stops = set(stopwords.words('english'))
stops = stops.union({
    'said','would','could','told','also','one','two','mr','new','year'
})
```

```python
df = pd.read_csv('../01-Vector-Model-And-Text-Processing/bbc_text_cls.csv')
df.head()
```

```python
vectorizor = CountVectorizer(stop_words=stops)
X = vectorizor.fit_transform(df['text'])
print(X.shape)
```

```python
noOfTopics = 5
lda = LatentDirichletAllocation(n_components=noOfTopics,random_state=12345)
lda.fit(X)
```

```python
# topics x wordds matrix.
print("No Of Topics --> ", lda.components_.shape[0])
print('No Of Words  -->' , lda.components_.shape[1])
print(' Matrix Size -->' , lda.components_.shape)
```

```python
def plot_top_words(model,features,n_top_words=10):
    fig,axes = plt.subplots(1,5,figsize=(30,15),sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_index = topic.argsort()[: -n_top_words -1 : -1]
        top_features = [features[i] for i in top_features_index]
        weights = topic[top_features_index]
        
        ax = axes[topic_idx]
        ax.barh(top_features, weights,height=0.7)
        ax.set_title(f"Topic {topic_idx + 1}",fontdict={"fontsize":30})
        ax.invert_yaxis()
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in " top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle('LDA', fontsize=30)
```

```python
features = vectorizor.get_feature_names_out()
plot_top_words(lda,features)
```

```python
# document x topics
Z = lda.transform(X)
```

```python
np.random.seed(0)
i = np.random.choice(len(df))
z = Z[i]
topics = np.arange(noOfTopics) + 1
fig, ax = plt.subplots()
ax.barh(topics,z)
ax.set_yticks(topics)
ax.set_title('True label: %s ' % df.iloc[i]['labels'])
```
