```python
import pandas as pd
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
df = pd.read_csv('../data/Tickets.csv')
df.head()
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
#Generating embeddings for our input dataset
def create_embeddings_from_dataframe(df,embeddings):
    df[2] = df[0].apply(lambda x: embeddings.embed_query(x))
    return df

#Read dataset for model creation
def read_data_set():
    df = pd.read_csv('../data/Tickets.csv', header=None)
    return df

#Splitting the data into train & test
def split_train_test_data(df_sample):
    # Split into training and testing sets
    sentences_train, sentences_test, labels_train, labels_test = train_test_split(
    list(df_sample[2]), list(df_sample[1]), test_size=0.25, random_state=0)
    print(len(sentences_train))
    return sentences_train, sentences_test, labels_train, labels_test
```

```python
df=read_data_set()
```

```python
clean_data = create_embeddings_from_dataframe(df,embeddings )
```

```python
labels_test = split_train_test_data(clean_data)
```

```python
sentences_train, sentences_test, labels_train, labels_test = labels_test
```

```python
svm_classifier = make_pipeline(StandardScaler(), SVC(class_weight='balanced')) 
```

```python
svm_classifier
```

```python
svm_classifier.fit(sentences_train, labels_train)
```

```python
#Get the accuracy score on test data
def get_score(svm_classifier,sentences_test,labels_test):
    score = svm_classifier.score(sentences_test, labels_test)
    return score
```

```python
get_score(svm_classifier, sentences_test, labels_test)
```
