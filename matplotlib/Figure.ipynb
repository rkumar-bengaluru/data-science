```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = 2 * x
fig = plt.figure();
axis = fig.add_axes([0,0,1,1])
axis.plot(x,y)
```

```python
a = np.linspace(0,10,20)
b = a**4
```

```python
a
```

```python
b
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0,0,1,1])
axis1.plot(a,b)
axis1.set_xlim(0,11)
axis1.set_ylim(0,8000)
axis1.set_xlabel('A')
axis1.set_ylabel('B')
axis1.set_title('Large Plot')

axis2 = fig.add_axes([0.1,0.5,0.3,0.3])
axis2.plot(a,b)
axis2.set_xlim(1,2)
axis2.set_ylim(0,50)
axis2.set_xlabel('A')
axis2.set_ylabel('B')
axis2.set_title('Zoom In')
```

```python
fig = plt.figure(figsize=(2,2),dpi=200)
axis1 = fig.add_axes([0,0,1,1])
axis1.plot(a,b)
fig.savefig('figure.png',bbox_inches='tight')
```
