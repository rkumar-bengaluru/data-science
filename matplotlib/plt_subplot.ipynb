```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = x *2
```

```python
fig,axis = plt.subplots(1,2)
axis[0].plot(x,y)
axis[0].set_xlabel('X')
axis[0].set_ylabel('Y')
axis[1].plot(a,b)
fig.subplots_adjust(hspace=0.2,wspace=0.5)
fig.savefig('subplot.png',bbox_inches='tight')
```
