```python
my_object_df = df.select_dtypes(include='object')
my_num_df = df.select_dtypes(exclude='object')
```

```python
df_object_dummies = pd.get_dummies(my_object_df,drop_first=True)
```

```python
final_df = pd.concat([my_num_df,df_object_dummies],axis=1)
```
