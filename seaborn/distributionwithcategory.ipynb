```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
# boxplot
# violinplot
# swarmplot
# boxenplot or letter-value-plot
```

```python
sns.boxplot(x='salary',y='level of education',data=df)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.boxplot(x='salary',y='level of education',data=df)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.boxplot(y='salary',x='level of education',data=df)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.violinplot(x='salary',data=df,y='division')
```

```python
sns.swarmplot(y='division',data=df,x='salary')
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.boxenplot(x='level of education',y='salary',data=df)
```

```python
df.groupby('level of education').mean()
```
