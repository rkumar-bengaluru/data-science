```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# jointplot
# pairplot
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df,hue='writing score')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

```python
sns.pairplot(data=df)
```

```python
df1 = df[['math score','reading score']]
df1.head()
sns.pairplot(df1)
```
