```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
df.head()
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df[['Birth rate','Mortality rate','Infant mortality rate','Growth rate']],lw=0.5,annot=True)
```

```python
sns.clustermap(df[['Birth rate','Mortality rate','Infant mortality rate','Growth rate']],lw=0.5,annot=True)
```
