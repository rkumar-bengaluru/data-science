```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/customerinformation_all.csv',converters={'kunnr': str})
```

```python
df.head()
```

```python
dfk = df[df['kunnr'] == '0030078914']
dfk.to_csv('0030078914.csv')
```

```python
kunnrgroup = df.groupby("kunnr")["kunn2"].count()
kunnrgroup = kunnrgroup.sort_values(ascending=False)
kunnrgroup = kunnrgroup.head(10)
kunnrgroup
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.barplot(x=kunnrgroup.index,y=kunnrgroup)
plt.xticks(rotation=90)
plt.xlabel('Customer Code')
plt.ylabel('ShipToCode')
plt.savefig('customercode_to_shiptocode_numbers.jpg')
```
