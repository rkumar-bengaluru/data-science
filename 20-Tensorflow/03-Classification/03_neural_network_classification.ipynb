# Decision Bounadry Function

* Take in model and features X and label y
* Create a meshgrid of the different X values
* Make Predictions across the mesh grid
* Plot the predictions as well as a line between zones (where each unique class falls)

```python
import numpy as np
# function to visualize decision boundary
def decision_boundary(model, X, y):
    """
    Plot the decision boundary created by a model predicting on X
    
    1. https://cs231n.github.io/neural-networks-case-study/
    2. 
    """
    # define the axis boundary of the plot and create a meshgrid
    x_min, x_max = X[:,0].min() - 0.1, X[:,0].max() + 0.1
    y_min, y_max = X[:,1].min() -0.1,  X[:,1].max() + 0.1
    
    xx,yy = np.meshgrid(
        np.linspace(x_min,x_max,100),
        np.linspace(y_min,y_max,100)
    )
    
    x_in = np.c_[xx.ravel(),yy.ravel()]
    y_pred = model.predict(x_in)
    print(y_pred)
    # check for multi-class
    if len(y_pred[0]) > 1:
        print('multiclass classification')
        y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        print('binary classification')
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.5)
    plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())      
```

```python
def plot_multiclass_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    
  
    
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    cmap = plt.cm.RdYlBu
    
    x_in = np.c_[xx.ravel(),yy.ravel()]
    y_pred = model.predict(x_in)
    
    
    #y_pred = y_pred.reshape(xx.shape)
    
     # check for multi-class
    if len(y_pred[0]) > 1:
        print('multiclass classification')
        y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        print('binary classification')
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    plt.contourf(xx, yy, y_pred, cmap=cmap, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

# Find the best learning rate

* Learning Rate callback 
* Another model 
* Modified loss curve plot

```python
# set random seed
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
# compile the model.
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])
# create callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))
# fit the model
history = model_1.fit(X_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
# check the history
pd.DataFrame(history.history).plot()
```

```python
# plot learning rate versus the loss
lrs = 1e-4 * 10**(tf.range(100)/20)
lrs[0],lrs[1]
```
